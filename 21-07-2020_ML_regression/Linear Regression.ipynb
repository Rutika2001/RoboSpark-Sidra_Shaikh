```python
x=df['Marketing Spend']
y=df['Profit']
m=len(x)
```

```python
def h(x, a0, a1):                             # hypothesis
    h = a0 + a1*x
    return h
def J(a0, a1, x, y, m):                       #cost function
    j = 0
    for i in range (m):
        j += (h(x[i], a0, a1) - y[i]) ** 2
    return j/(2*m)
def gradJ(a0, a1, x, y, m):                    # gradient of cost function
    Dja0 = 0 
    Dja1 = 0
    for i in range (m):
        Dja0 += (h(x[i],a0,a1) - y[i])
        Dja1 += (h(x[i],a0,a1) - y[i]) * x[i]
    gradJ = [Dja0/m, Dja1/m]
    return gradJ
def ap1(a0, a1, gJ, alpha):                    # updating values of theta
    a0p1 = a0 - alpha * gJ[0]
    a1p1 = a1 - alpha * gJ[1]
    ap1 = [a0p1, a1p1]
    return ap1
```

```python
fig, ax = plt.subplots()
p0 = np.linspace(-10,10, 500)
p1 = np.linspace(-10,10, 500)
pa0, pa1 = np.meshgrid(p0, p1)      
ax.contour(pa0,pa1,J(pa0,pa1,x/500000,y/500000,m=len(x)) , colors='C0', levels=[i for i in np.arange(0,60,2)])
a0 = 6
a1 = 2.2
ax.plot(a0,a1, 'o')

for i in range(300):
    a0old, a1old = a0, a1
    a0,a1 = ap1(a0,a1, gJ=gradJ(a0,a1,x/500000,y/500000,m), alpha=0.1)
    ax.plot(a0,a1,'o')
    ax.plot([a0old,a0],[a1old,a1], '-')
    

print(a0,a1)
```

```python
x_mean = x.mean()
x_std = x.std()

y_mean = y.mean()
y_std = y.std()

x_new = (x - x_mean) / x_std           #mean Normalization
y_new = (y - y_mean) / y_std
```

```python
def plot_s(start, alpha, epoch):
    fig, ax = plt.subplots()
    p0 = np.linspace(-6,6, 500)
    p1 = np.linspace(-6,6, 500)
    pa0, pa1 = np.meshgrid(p0, p1)

    ax.contour(pa0,pa1,J(pa0,pa1,x_new,y_new,m=len(x)) , colors='C0',levels=[i for i in np.arange(0,80,)])

    a0 = start[0]
    a1 = start[1]
    ax.plot(a0,a1, 'o')

    for i in range(epoch):
        a0old, a1old = a0, a1
        a0,a1 = ap1(a0,a1, gJ=gradJ(a0,a1,x_new,y_new,m), alpha=alpha )
        ax.plot(a0,a1,'o')
        ax.plot([a0old,a0],[a1old,a1], '-')
```

```python
plot_s([-1,4], 0.5, 10)
```

```python
def plot_graph(x, y, a0, a1):
    fig, ax = plt.subplots()
    ax.plot(x, y, '*', label = "traning data")
    ax.plot(x, h(x_new, a0, a1), label = "h(x) = %.2f + %.2f * x" %(a0, a1))
    plt.legend()
```

```python
a0 = -1
a1 =4
for i in range (10) :
    a0, a1 = ap1(a0, a1, gJ = gradJ(a0, a1, x_new, y_new, m), alpha = 0.35)
    plot_graph(x_new, y_new, a0, a1)
```

```python
def predict(x, a0, a1) :                            
    pre = y_std * (a0 + a1 * (x - x_mean)/x_std) + y_mean
    return pre
```
